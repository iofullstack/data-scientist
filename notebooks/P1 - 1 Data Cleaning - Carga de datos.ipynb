### Ejercicio de descarga de datos con urllib3

Realizar un ejercicio usando la librería `urllib3` para leer los datos desde una URL externa, procesarlos y convertirlos a un data frame de python antes de guardarlos a un `CSV` local.

````python
def downloadFromUrl(url, filename, mainpath="/Users/ggary/anaconda_projects/datasets", sep=",", delim="\n", encoding="utf-8"):
    import pandas as pd
    import urllib3
    http = urllib3.PoolManager()
    r = http.request('GET', url)
    print("El estado de la respuesta es %d" %(r.status))
    response = r.data
    
    str_data = response.decode(encoding)
    lines = str_data.split(delim)
    col_names = lines[0].split(sep)
    n_cols = len(col_names)
    
    counter = 0
    main_dict = {}
    for col in col_names:
        main_dict[col] = []
    
    for line in lines:
        if (counter > 0):
            values_quotes = line.strip().replace(", ", "; ")
            values = values_quotes.strip().split(sep)
            for i in range(n_col):
                if ('; ' in values[i]):
                    values[i] = values[i].replace("; ", " ")
                main_dict[col_names[i]].append(values[i])
        counter += 1
    
    print("El data set tiene %d filas y %d columnas" %(counter-1, n_cols))
    
    df = pd.DataFrame(main_dict)
    fullpath = os.path.join(mainpath, filename)
    
    df.to_csv(fullpath + ".csv")
    df.to_json(fullpath + ".json")
    df.to_excel(fullpath + ".xlsx")
    print("Los ficheros se han guardado correctamente en: " + fullpath)
    
    return df
````
